# <b> Dealing With Missing Or Null Values : 

```python
# Getting the missing values, also showing sum of null values for each column in a sorted manner
dataframe.isnull().sum().sort_values(ascending=True)
```

```python
# Getting total null value in percent that each columns contains
dataframe.isnull().sum()*100/len(dataframe)
```

### So, we find here that <i><b> company </b> </i> contains <b> 94 % </b> and <i><b>agent</b> </i> contains <b> 13.68 % </b> null values here so let's drop these columns as they wont give much of a info and replace </b> </i> <b> <i> country</b> and <b> children </b> </i>columns null value with <b> mode </b> and <b>mean</b> respectively.

```python
# Dropping the column company and agent from the dataframe
dataframe.drop(['company','agent'],inplace=True,axis=1)

# Replace the null values of country column with mode value
dataframe['country'].fillna(dataframe.country.mode().to_string(), inplace=True)

# Replace the null values of children column with mean value
dataframe['children'].fillna(round(dataframe.children.mean()), inplace=True)

# displaying dataframe
dataframe.head()
```

```python
# Let's now check for null or missing value in our dataframe
dataframe.isnull().sum()
```

## <b> <i> So, there is no null or missing value left in the dataframe. </b> </i>

## <b> 1. How booking of hotels differs year wise ?

```python
# Let's see in what is booking distribution year wise
year_wise_dis = dataframe['arrival_date_year'].value_counts()
year_wise_dis
```

```python
# Let's see above data graphically using count plot
plt.figure(figsize=(7,5))
sns.countplot(data=dataframe, x='arrival_date_year')
plt.title("Booking Distribution of Hotels Year Wise ", fontsize = 22)
plt.ylabel('Number Of Bookings',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
plt.show()
```

## <b>Comparison in percentage : 

```python
# Now let see booking ratio of hotels year wise
year_wise_booking_ratio = dataframe["arrival_date_year"].value_counts()
year_wise_booking_ratio.plot.pie(figsize = (12,6),fontsize = 16, autopct="%.1f%%")
plt.title('Year wise Booking Ratio of Hotels in percentage',fontsize = 22)
plt.ylabel('Bookings Percentage',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
pass
```

#### So, after seeing above visuals and above output data we can say that the year <b>2016</b> has highest number of booking out of other three years. It has total number of <b>56707 [47.5 %] </b> bookings whereas <b>2017</b> is second in numbers of bookings and have <b>40687 [34.1 %]</b> total bookings and <b>2015</b> have least <b>21996 [18.4 %]</b> bookings.

## <b>2. What is the hotel booking difference between Resort Hotel and City Hotel ?

```python
# Let's see the booking difference between city and resort hotel
booking_diff_of_hotels = dataframe['hotel'].value_counts()
booking_diff_of_hotels
```

```python
# Let's compare booking difference between city and resort hotel on plot
plt.figure(figsize=(8,5))
sns.countplot(data=dataframe, x = 'hotel')
plt.title("Booking Difference Between City And Resort Hotel", fontsize = 22)
plt.ylabel('Number Of Bookings',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
plt.show()
```

#### So, after seeing visuals and above output data we can say that <b>City Hotel</b> is booked approximately twice then <b>Resort Hotel</b>. Total number of bookings in <b>City Hotel</b> is <b><u> 79330</u> </b> whereas total number of bookings in <b>Resort Hotel</b> is <b><u> 40060</u> </b>

# <b> Dealing With Missing Or Null Values : 

```python
# Getting the missing values, also showing sum of null values for each column in a sorted manner
dataframe.isnull().sum().sort_values(ascending=True)
```

```python
# Getting total null value in percent that each columns contains
dataframe.isnull().sum()*100/len(dataframe)
```

### So, we find here that <i><b> company </b> </i> contains <b> 94.30 % </b> and <i><b>agent</b> </i> contains <b> 13.68 % </b> null values here so let's drop these columns as they wont give much of a info and replace </b> </i> <b> <i> country</b> and <b> children </b> </i>columns null value with <b> mode </b> and <b>mean</b> respectively.

```python
# Dropping the column company and agent from the dataframe
dataframe.drop(['company','agent'],inplace=True,axis=1)

# Replace the null values of country column with mode value
dataframe['country'].fillna(dataframe.country.mode().to_string(), inplace=True)

# Replace the null values of children column with mean value
dataframe['children'].fillna(round(dataframe.children.mean()), inplace=True)

# displaying dataframe
dataframe.head()
```

## <b> 1. How booking of hotels differs year wise ?

```python
# Let's see above data graphically using count plot
plt.figure(figsize=(7,5))
sns.countplot(data=dataframe, x='arrival_date_year')
plt.title("Booking Distribution of Hotels Year Wise ", fontsize = 22)
plt.ylabel('Number Of Bookings',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
plt.show()
```

## <b>2.  In which month most of booking was done for both City Hotel and Resort Hotel seperately ?

```python
# First, we use lineplot to to display the monthly trend for both City and Resort Hotel simultaneously.

total_bookings = dataframe["arrival_date_month"].value_counts()
total_bookings.plot.pie(figsize = (14,6),fontsize = 12, autopct="%.1f%%")
plt.title('Month Wise Booking Distribution For Both  City Hotel and Resort Hotel Simultaneously ',fontsize = 24)
plt.ylabel('Bookings In %',fontsize = 24)
plt.xlabel('Month Of Bookings',fontsize = 22)
pass
```

### So, in <b> August</b> booking is mostly higher <b> 11.6 </b> on the basis of month wise booking distribution for both  City Hotel and Resort Hotel simultaneously. 

```python
# So we need to split the data in 2 sets for City And Resort Hotel respectively so we can easily understand trends for City and Resort Hotel seperately.

city_hotel = dataframe[dataframe['hotel']=='City Hotel']
resort_hotel = dataframe[dataframe['hotel']=='Resort Hotel']
```

```python
# Just because Bookings in 2016 Was Higher then other years so to check the distribution of booking on monthly basis 
# we will take only the year 2016 in consideration for both hotel and resort 

city_hotel_in_2016 = city_hotel[city_hotel['arrival_date_year']==2016]
resort_hotel_in_2016 = resort_hotel[resort_hotel['arrival_date_year']==2016]

fig, axes = plt.subplots(1, 2, figsize=(22, 6))
fig.suptitle('Distribution Of Monthwise Bookings  for City and Resort Hotel in 2016', fontsize=20)

# Monthwise Bookings Distribution for Resort Hotel in 2016
sns.countplot(ax = axes[0], data = city_hotel_in_2016, x = "arrival_date_month")
axes[0].set_title('Monthwise Bookings Distribution for City Hotel in 2016', fontsize=15)

# Monthwise Bookings Distribution for Resort Hotel in 2016
sns.countplot(ax = axes[1], data=resort_hotel_in_2016, x= "arrival_date_month" )
axes[1].set_title('Monthwise Bookings Distribution for Resort Hotel in 2016', fontsize=15)
plt.show()
```

### <b> We can see that in both the cases October shows higest montly bookings being done and lowest done in January. Hence, its shows most of guest take interest to visit hotel in month of october.

## <b>3. What is the hotel booking difference between Resort Hotel and City Hotel ?

## <b> 4. What is difference in Cancelled and Not Cancelled Hotel Booking ? Also check it for both city and resort hotel ?

## <b>Comparison in percentage : 

```python
# Let's see how many bookings is cancelled and not cancelled in percent(%)
plt.figure(figsize=(12,6))
status = [dataframe.is_canceled.sum(),len(dataframe)-dataframe.is_canceled.sum()]
labels=['Canceled Booking In %','Not Canceled Bookings In %']
plt.pie(status, labels=labels, autopct='%.0f%%')
plt.title('Comparison of  Cancelled and Not Cancelled Bookings In Percent (%)', fontsize=22)
plt.show()
```

### So, we saw here that almost <b>37%</b> bookings were cancelled whereas approximately <b>67%</b> booking were successful and not canceled .

### <b> <i>Now, let's see how cancelled and not cancelled booking distributed for both City And Resort Hotel 

```python
# cancelled and not cancelled booking distributed for City Hotel And Resort Hotel
plt.figure(figsize=(9,6))
sns.countplot(data=dataframe, x='hotel', hue='is_canceled')
plt.title("Cancelled and Not Cancelled Bookings Distributed For Both City And Resort Hotel", fontsize = 22)
plt.xlabel("Hotel", fontsize = 15)
plt.ylabel("Number Of Bookings", fontsize = 15)
plt.show()
```

##<b> Firstly, here " 0 " represent bookings which are "<u> Not Canceled</u> " whereas " 1 " represent bookings which is "<u> Canceled</u> ".

---

#### <b>*   We can see here that there is huge number of bookings were cancelled in case of <i> City Hotel </i> even though number of bookings for <i> City Hotel </i> is higher in comparison to <I> Resort Hotel </i> whereas for <i> Resort Hotel </i> number of cancellation is comparatively low in comparison to <i> City Hotel </i>.


*   ### <b> Most probably, people books <i> Resort Hotel </i> for spending their free times with their family and friends or also for family parties which mostly people not cancel but peoples book <i> City Hotel </i> mostly for business meetings, and some same reasons which increase the chances of cancellation of booking as many times meetings cancelled.

## <b> 5. From which top 7 countries mostly(maximum) guests comes in both hotels seperately ?

```python
# Let's see from which top 7 countries mostly guests comes or bookings done

fig, axes = plt.subplots(1, 2, figsize=(22, 7))
fig.suptitle('Distribution Of Countries on the basis Of mostly(maximum) guests comes in both hotels(City And Resort)', fontsize=20)
sns.countplot(ax = axes[0], data = city_hotel_in_2016, x = "country", order = city_hotel_in_2016.country.value_counts().iloc[:7].index)
axes[0].set_title('Distribution Of Countries on the basis of maximum guests comes in City Hotel', fontsize=15)
sns.countplot(ax = axes[1], data=resort_hotel_in_2016, x= "country", order = resort_hotel_in_2016.country.value_counts().iloc[:7].index )
axes[1].set_title('Distribution Of Countries on the basis of maximum guests comes in Resort Hotel', fontsize=15)
plt.show()
```

#### Above data shows that the maximum guests booked the hotels (in case of both City and Resort Hotel) is belong from country <b>Portugal (PRT)</b>. Most probably, after seeing above visuals it is mostly understandable that this dataset shows the data of hotels which located in <b>Portugal.</b>

### <b>Portugal(PRT), France(FRA), Great Britain(GRB), Spain(ESP) and Germany(DEU)</b> are the top countries from most guests come, more than 80% come from these 5 countries.
